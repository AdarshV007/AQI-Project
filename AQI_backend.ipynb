```python
import numpy as np
```

```python
import pandas as pd
```

```python
data=pd.read_csv(r'C:\Users\ASUS\Downloads\YEARLY_2023_CAAQMS_FINAL_UPLOAD_merged (1).csv',na_values='-')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
x=data[['CO','Ozone','NO2','SO2','PM2.5','PM10']]
y=data['AQI']
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
reg1=KNeighborsRegressor(n_neighbors=3)
```

```python
reg1.fit(x_train,y_train)
```

```python
y_pred=reg1.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model=DecisionTreeRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.ensemble import RandomForestRegressor
reg2=RandomForestRegressor(n_estimators=100,random_state=0)
reg2.fit(x_train,y_train)
```

```python
y_pred=reg2.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg3=LinearRegression()
```

```python
reg3.fit(x_train,y_train)
```

```python
y_pred=reg3.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
```
